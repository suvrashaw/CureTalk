```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestClassifier 
import pickle
```

```python
dataset = pd.read_csv("dataset.csv")
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
len(dataset["Disease"].value_counts())
```

```python
cols = dataset.columns
data = dataset[cols].values.flatten()

s = pd.Series(data)
s = s.str.strip()
s = s.values.reshape(dataset.shape)

dataset = pd.DataFrame(s, columns=dataset.columns)
dataset.head()
```

```python
dataset = dataset.fillna(0)
dataset.head()
```

```python
symptom_severity = pd.read_csv("symptom_severity.csv")
symptom_severity.head()
```

```python
symptom_list = list(symptom_severity["Symptom"])
symptom_list
```

```python
dataset_1 = pd.DataFrame(columns = symptom_list)

dataset_1["Disease"] = dataset["Disease"]
first_column = dataset_1.pop('Disease')
dataset_1.insert(0, 'Disease', first_column)

dataset_1 = dataset_1.fillna(0)
dataset_1
```

```python
%%time
for i in range(len(dataset)):
    for ele in list(dataset.columns):
        symp = dataset[ele][i]
        for elem in list(dataset_1.columns):
            if(elem == symp):
                dataset_1.loc[i, elem] = 1 
                break
                
dataset_1
```

```python
%%time
symptom_severity_value_list = list(symptom_severity["weight"])
i=0
for ele in symptom_list:
    dataset_1[ele] = dataset_1[ele] * symptom_severity_value_list[i]
    i=i+1

dataset_1
```

```python
X_train = dataset_1.iloc[:,1:].values
y_train = dataset_1['Disease'].values

X_train[100]
```

```python
print(y_train[100])
```

```python
# creating a RF classifier
model = RandomForestClassifier(n_estimators = 100)  
  
# Training the model on the training dataset
model.fit(X_train, y_train)
```
